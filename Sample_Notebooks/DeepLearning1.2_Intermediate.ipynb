# VECTORS

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.quiver(0,0,7,8, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(-2,-3,-4,-4,scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.plot()
```

```
def ploty(v):
  colors=['r','g','b']
  c=0
  for i in v:
    plt.quiver(i[0],i[1],i[2],i[3],scale_units='xy', angles='xy', scale=2, color=colors[c%3])
    c+=1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.plot()
```

```
ploty([(0,0,3,4),(0,0,-3,-3),(0,0,-3,4),(0,-1,4,-1)])
```

```
vecs=[(0,0,3,4),(0,0,-3,-3),(0,0,-3,4),(0,-1,4,-1)]
```

```
vectors=list(map(lambda x: np.asarray(x), vecs))
print(vectors)
```

```
ploty(vectors)
```

```
vectorsadded=vectors[1]+vectors[2]
```

```
ploty(vectorsadded)
```
